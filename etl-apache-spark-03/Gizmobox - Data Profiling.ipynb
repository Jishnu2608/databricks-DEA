
## Profile Data manually
### count function

```sql
select count(*), count(customer_id), count(email), count(telephone) from gizmobox.bronze.v_customers;
```


### count_if function

```sql
select count(*), count_if(customer_id is not null), count_if(email is not null), count_if(telephone is not null) from gizmobox.bronze.v_customers;
```


### where function

```sql
select count(*) from gizmobox.bronze.v_customers where customer_id is null;
```

### distinct keyword

```sql
select count(*)total_number_of_records, 
count(distinct customer_id) unique_customer_ids
from gizmobox.bronze.v_customers where customer_id is not null;
```
